## **Three diffrerent types of Machine Learning**



1 **Supervised Learning**
*   The model learns from labeled data.
*   **Goal:** Predict outcomes for new data.
*   **Examples:** Classification, Regression (house price prediction).

2 **Unsupervised Learning**
*   The model learns patterns from unlabeled data.
*   **Goal:** Discover structure in the data.
*   **Examples:** Clustering (customer segmentation), Dimensionality Reduction (PCA , VAE).

3 **Reinforcement Learning**


*   The model learns by interacting with an environment and receiving rewards or penalties.
*   **Goal**: Optimize actions through trial and error.
*   **Examples**: Robotics, game-playing AI.


## **Classification VS Regression**

### **Classification**

Classification is a type of supervised learning where the goal is to predict a label or category for each input.
The output is discrete (fixed classes).
Examples:


*   Email  Spam / Not Spam
*   Image  Cat / Dog
*   Patient  Healthy / Sick

**Key idea**: The model learns decision boundaries to assign correct classes.

### **Regression**

Regression is a type of supervised learning where the goal is to predict a continuous numerical value.
The output can take any real number.
Examples:
*   Predicting house prices
*   Estimating blood pressure value
*   Forecasting temperature

**Key idea**: The model learns patterns to estimate numeric outcomes.

# **Unsupervised Learning**

Unsupervised Learning is a type of Machine Learning where the model learns patterns from data without labeled outputs.

 **There is no target variable** > the algorithm discovers hidden structure on its own.

### ** Main Applications**
#### **1 Dimension Reduction**

Used when datasets have many features, making them hard to visualize or train on.

**Goal:**


*   Keep the most important information
*   Reduce the number of features

**Common methods:**



*   PCA (Principal Component Analysis)
*   t-SNE (t-Distributed Stochastic Neighbor Embedding)
*   Variational Autoencoder (VAE)


** Example:** Thousands of gene expression features  compress to 23 dimensions to visualize patterns



#### **2 Finding Subgroups with Clustering**

Used to group similar data points without telling the model what the groups are.

**Goal:**

*   Discover hidden substructures

*   Identify segments or subpopulations

**Common methods:**

*   K-means clustering



** Example:** Group patients into subtypes based on clinical data
 may reveal disease subgroups with different treatment responses


# **Terminology**
In Machine Learning, our goal is to build a model that can learn patterns from data and make predictions. To do that, we separate the data into:

** Target Variable** (Label / Output / Dependent Variable)

This is what we want the model to predict

Example: Predicting house price  Price = Target


** Features** (Predictors / Inputs / Independent Variables)

The information we use to predict the target

Can be many features and each feature gives useful signals to the model

Example: House size, number of rooms, location  Features


** Sample**( Instance / Observation / Data Point)

One row of data from the dataset representing one item being studied.

Example:One house in a housing dataset  One sample

** Class Label**

(Used in classification problems)

The category or class assigned to each sample.

Example:Spam vs Not Spam  2 class labels
